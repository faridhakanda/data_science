## Data Science and Web development

```python
import seaborn as sns
import pandas as pd
sns.set_theme()
tips = sns.load_dataset("tips")
sns.relplot(
    data = tips, x="total_bill", y="tip", col="time", hue="smoker", style="smoker", size="size"
)
dots = sns.load_dataset("dots")
sns.relplot(
    data = dots, kind="line", x="time", y="firing_rate", col="align", hue="choice", size="coherence", style="choice", facet_kws=dict(sharex=False)
)
```
